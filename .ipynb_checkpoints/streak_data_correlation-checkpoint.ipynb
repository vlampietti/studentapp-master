```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
streak_data = pd.read_csv('data/streak_data.csv')
```

```python
def display_corr(columns, df):
    cols = columns
    if (cols is None):
        cols = list(df)
        
    fig = plt.figure(figsize=(9,9))
    ax = fig.add_subplot(111)
    cax = ax.matshow(df[cols].corr(), interpolation='nearest')
    fig.colorbar(cax)

    ax.set_xticklabels(['']+cols)
    ax.set_yticklabels(['']+cols)
    
    ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
    
    plt.show()
```

```python
columns = ['2X','3X','4X','5X','6X','7X','8X','9X','10X']
display_corr(columns, streak_data)
```

```python
def display(columns, df):
    df.reset_index().plot(x='streak_data',y=columns, figsize=(15, 6))
```

```python
total = streak_data.sum().reset_index()
display(['2X','3X','4x','5','6X','7X','8X','9X','10X'],total)
```
